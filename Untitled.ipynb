```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
from mplsoccer import Pitch
```

```python
# ---------- Step 1: Load data from GitHub ----------
# Example match (Barcelona vs Deportivo Alavés, La Liga 2017-18)
url = "https://raw.githubusercontent.com/statsbomb/open-data/master/data/events/7561.json"
events_json = requests.get(url).json()

events = pd.json_normalize(events_json, sep="_")

# ---------- Step 2: Clean data ----------
# keep only rows with location + player
events = events.dropna(subset=["location", "player_name", "team_name"])

# expand location into x, y
events["x"] = events["location"].apply(lambda loc: loc[0] if isinstance(loc, list) else None)
events["y"] = events["location"].apply(lambda loc: loc[1] if isinstance(loc, list) else None)

# normalize pitch (StatsBomb uses 120x80 → scale to 0–1)
events["x_norm"] = events["x"] / 120
events["y_norm"] = events["y"] / 80

# ---------- Step 3: Average positions ----------
avg_positions = (
    events.groupby(["team_name", "player_name"])
    .agg({"x": "mean", "y": "mean"})
    .reset_index()
)

# ---------- Step 4: Plot for one team ----------
team = avg_positions["team_name"].unique()[0]  # first team
team_positions = avg_positions[avg_positions["team_name"] == team]

pitch = Pitch(pitch_type="statsbomb", line_color="black")
fig, ax = pitch.draw(figsize=(10, 7))

pitch.scatter(
    team_positions["x"], team_positions["y"],
    s=300, c="red", edgecolors="black", ax=ax
)

for i, row in team_positions.iterrows():
    ax.text(row["x"], row["y"]+2, row["player_name"], ha="center", fontsize=8)

plt.title(f"Average Formation - {team}", fontsize=14)
plt.show()
```

```python
fig, axs = pitch.draw(figsize=(16, 10), ncols=2)
for i, team in enumerate(avg_positions["team_name"].unique()):
    team_positions = avg_positions[avg_positions["team_name"] == team]
    pitch.scatter(team_positions["x"], team_positions["y"], 
                  s=300, c="red" if i==0 else "blue", 
                  edgecolors="black", ax=axs[i])
    for _, row in team_positions.iterrows():
        axs[i].text(row["x"], row["y"]+2, row["player_name"], ha="center", fontsize=7)
    axs[i].set_title(f"Average Formation - {team}")
plt.show()
```

```python
from mplsoccer import VerticalPitch

pitch = VerticalPitch(pitch_type="statsbomb", line_color="black")
fig, ax = pitch.draw(figsize=(7, 10))

player = "Lionel Andrés Messi Cuccittini"
messidata = events[events["player_name"] == player]

pitch.kdeplot(messidata["x"], messidata["y"], 
              ax=ax, cmap="Reds", fill=True, n_levels=50)
ax.set_title(f"Heatmap - {player}")
plt.show()
```
